## Persistence

Most of the code we have run so far are transient. They will run during your session and after producing some output, their data disappears.

Other programs need to be **persistent**, meaning that they need to store (at least some of) their data.

This chapter will focus on reading and writing files to safe data.

## Opening a file

You can open a file with `open()`

You can write a to a file using `'w'` as a second parameter: `open('filename.txt', 'w')`

If the file exists, running this command will replace the file. If the file does not exist, it will be created.

```python
fout = open('output.txt', 'w')
```

The function `open()` returns a file object that allows you to write to the file.

```python
line1 = "This is the first line of text.\n"
fout.write(line1)
```

The method returns the number of characters that were written.

If you call write again, it adds the new data to the end of the file.

```python
line2 = "This will be the second line of text.\n"
fout.write(line2)
```

When you are done writing, you should close the file.

```python
fout.close()
```

Closing the file frees up the system resources dedicated to tracking that file. If you forget to close the file and Python sees no object names bound to the file object, Python's garbage collector will automatically close the connection.

## with

You can use the `with` keyword to have operations run with a particular file. When the operations finish, Python will automatically close the file. 

```python
# opening a file and writing lines to it:
with open('output.txt', 'w') as file:
    file.write(line1)
    file.write(line2)
```

```python
# opening the file and printing each line
with open('output.txt', 'r') as file:
    for i in file:
        print(i)
```

## The Format Operator

When using `write`, the argument has to be a string.

The easiest way to convert non-strings to strings is with `str()`. If you don't convert the value to a string, you can run into problems.

```python
x = 42.3
y = x + " is my favorite number"
```

```python
x = 42.3
y = str(x) + " is my favorite number"
print(y)
```

Another option is to use the format method and curly brace `{}` replacement fields as placeholders for the variables to be inserted into the string.

Python has a **format specification mini language** which is used to specify the format of the string.

See: https://docs.python.org/3/library/string.html#format-string-syntax

Strings with placeholders can also be filled in-line with f-strings. Both f-strings and str.format() insert values into strings using the same format mini-language.

```python
x = 12.34
y = '{} is my favorite number'.format(x)
print(y)
```

```python
# equivalent f-string
# when using the f-string, you must place the object name inside the text placeholder
yfstring = f'{x} is my favorite number'
print(yfstring)
```

```python
x = 12.34
x2 = 37.8
# you can use multiple replacement fields and python will insert value positionally
y = '{} is my favorite number and {} is my second favorite number'.format(x, x2)
print(y)
```

```python
x = 12.34
x2 = 37.8
# you can set index values for replacement fields and python will insert values
# based on the index of the tuple given to .format( )
y = '''{1} is my favorite number and {0} is my second favorite number.
Again, my favorite number is {1}.'''.format(x, x2)
print(y)
```

```python
x = 321
x2 = 567
yfstring = f'''{x} is my favorite number and {x2} is my second favorite number. 
Again, my favorite number is {x}.'''
print(yfstring)
```

```python
# You can add :g in the replacement field.
# This prints floating values in general format
# Default is round to 6 significant figures.
# You can specify my signficant figures with .8g for 8 significant figures.
x  = 0.000123456789
x2 = 0.0000123456789
y  = '{0:g} is my favorite number. I also like {1:.8g}.'.format(x, x2)
print(y)
```

```python
# f-strings use the same python text format specifications
yfstring  = f'{x:g} is my favorite number. I also like {x2:.8g}.'
print(yfstring)
```

```python
# :f in the replacement field prints in fixed point format.
# This prints floating values in fixed point format, default 6 fixed places after the decimal.
# If you write .8f then it will provide 8 places after the decimal
x = 0.000123456789
x2 = 0.0000123456789
y = '{0:f} is my favorite number. I also like {1:.8f}'.format(x, x2)
print(y)
```

```python
# alternative method using + operator
x = 12.345
age = 85
y = 'My age is ' + str(age) + ', and ' + str(x) + ' is my favorite number'
print(y)
```

```python
# placeholder indices are optional.
y = 'My name is {:s}, I am {:g} years old, and {:g} is my favorite number'.format("Joe Bruin", 100, 83)
print(y)
```

### common f-strings

```python
# Money with 2 decimals
amount = 1234.5
print(f"${amount:,.2f}")
```

```python
# Percentages
ratio = 0.0739
print(f"{ratio:.1%}")
```

```python
# Zero-padded integers
i = 42
print(f"{i:05d}")
```

```python
# hours:minutes:seconds ensuring 2 charactes for each value
h = 1
m = 2
s = 7
d = "PM"
print(f'{h:02d}:{m:02d}:{s:02d} {d}')
```

```python
# Columnar table built from tuples
rows = [("penguin", 3), ("albatross", 12)]
for animal, n in rows:
    print(f"{animal:<12} {n:>5d}")
```

## Catching Exceptions

When dealing with files, and also just in programming in general, a lot of things can go wrong and Python will throw an exception.

You can often catch many of the exceptions with `try ... except` which works in a manner similar to `if ... else`

```python
fin = open('bad_file')
```

```python
try:
    fin = open('bad_file')
except FileNotFoundError:
    print('The file was not found, but the exception was caught.')
```

Python first tries the commands in the `try:` block. If everything goes well, it skips the commands in the `except:` block. If it encounters an exception in the `try` block, it immediately exits the block and executes the code in the `except` block.


```python
try:
    fin = open('bad_file')
except KeyError:
    print('This will catch KeyErrors.') # because the exception is not a KeyError, this block will not execute
```

## Saving Python Objects with JSON

If you want to save Python objects like dictionaries or lists to a file so you can use them later, the modern standard is **JSON** (JavaScript Object Notation). 

Unlike `write()`, which only accepts strings, the `json` module translates Python objects into a text format that can be saved, and then translates them back into Python objects when loaded.

```python
import json

# Let's say we have a dictionary we want to save
my_data = {
    "name": "Joe Bruin",
    "age": 100
}
```

```python
my_data
```

```python
# We use json.dump() to write the object to a file
with open('data.json', 'w') as file:
    json.dump(my_data, file)
```

Now that the data is saved in `data.json`, we can retrieve it later using `json.load()`.

```python
# We use json.load() to read the file back into a Python object
with open('data.json', 'r') as file:
    loaded_data = json.load(file)

print(type(loaded_data))
print(loaded_data)
```

```python
my_data
```

```python
# If you want to save many objects to the same file, you must create a dictionary that contains
# all of the objects you want to save, and then write that dictionary to the file.
my_list = [1, 2, 3, 4, 5]
my_string = "This is a string."
my_number = 42

# Package everything into a single dictionary
data_to_save = {
    "my_data": my_data,
    "my_list": my_list,
    "my_string": my_string,
    "my_number": my_number
}

# Save that one dictionary to the file
with open('data.json', 'w') as file:
    # Adding indent=4 makes the file easily readable for humans!
    json.dump(data_to_save, file, indent=4)
```

```python
# when you are ready to read the file back, use json.load()
# to read the entire dictionary back into memory, and then pull out the variables you want using their keys
with open('data.json', 'r') as file:
    # This loads the entire dictionary back into memory
    loaded_data = json.load(file)

print(loaded_data)

# Now you can just pull your variables back out using their keys
loaded_dict = loaded_data["my_data"]
loaded_list = loaded_data["my_list"]
loaded_string = loaded_data["my_string"]
loaded_number = loaded_data["my_number"]
```

If you are dealing with a massive amount of data and you don't want to load a giant dictionary into memory all at once, you can use a format called **JSON Lines** (usually saved as `.jsonl`). 

In this format, you write exactly one valid JSON object per line. You have to convert the object to a string with `json.dumps()` (note the "s" for string) and manually add a newline character.

```python
list2 = [[1,2,3], "This is a string."]
```

```python
with open('data.jsonl', 'w') as file:
    # Write each object as a JSON string on its own line
    file.write(json.dumps(my_list) + '\n')
    file.write(json.dumps(list2) + '\n')
    file.write(json.dumps(my_string) + '\n')
    file.write(json.dumps(my_number) + '\n')
```

```python
with open('data.jsonl', 'r') as file:
    # Read the file line by line and parse each one
    loaded_item1 = json.loads(file.readline())
    loaded_item2 = json.loads(file.readline())
    loaded_item3 = json.loads(file.readline())
```

```python
loaded_item1
```

```python
# of course, you might not know how many lines are in the file, so you could read the file line by line
# in a loop until you reach the end of the file
# you could use a list to store the loaded items, and then append each loaded item to the list as you read it in
loaded_items = []
with open('data.jsonl', 'r') as file:
    for line in file:
        loaded_item = json.loads(line)
        loaded_items.append(loaded_item)
for item in loaded_items:
    print(item)
```
